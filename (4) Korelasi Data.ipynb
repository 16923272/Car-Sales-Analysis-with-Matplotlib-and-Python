```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel("car_sales_data.xlsx")
```

```python
# Korelasi antara date dengan car year
print('Nilai korelasi antara waktu pembelian mobil dengan tahun rilis mobil adalah:', data["Date"].corr(data["Car_Year"]))
# Korelasi antara date dengan sale price
print('Nilai korelasi antara waktu pembelian mobil dengan harga jual mobil adalah:', data["Date"].corr(data["Sale_Price"]))
# Korelasi antara date dengan commision rate
print('Nilai korelasi antara waktu pembelian mobil dengan tingkat komisi yang dibayarkan kepada penjual adalah:', data["Date"].corr(data["Commission_Rate"]))
# Korelasi antara date dengan commission earned
print('Nilai korelasi antara waktu pembelian mobil dengan tingkat komisi yang dibayarkan kepada penjual adalah:', data["Date"].corr(data["Commission_Earned"]))
```

```python
# Korelasi date
data.plot(kind="scatter", x="Date", y="Car_Year", xlabel="Date", ylabel="Car Year", title="Diagram Tebar Date dan Car Year")
data.plot(kind="scatter", x="Date", y="Sale_Price", xlabel="Date", ylabel="Sale Price", title="Diagram Tebar Date dan Sale Price")
data.plot(kind="scatter", x="Date", y="Commission_Rate", xlabel="Date", ylabel="Commission Rate", title="Diagram Tebar Date Commission Rate")
data.plot(kind="scatter", x="Date", y="Commission_Earned", xlabel="Date", ylabel="Commission Earned", title="Diagram Tebar Date dan Commission Earned")
```

```python
# Korelasi antara car year dengan date
print('Nilai korelasi antara tahun rilis mobil dengan tanggal pembelian mobil adalah:', data["Car_Year"].corr(data["Date"]))
# Korelasi antara car year dengan sale price
print('Nilai korelasi antara tahun rilis mobil dengan harga jual mobil adalah:', data["Car_Year"].corr(data["Sale_Price"]))
# Korelasi antara car year dengan commission rate
print('Nilai korelasi antara tahun rilis mobil dengan tingkat komisi yang dibayarkan kepada penjual adalah:', data["Car_Year"].corr(data["Commission_Rate"]))
# Korelasi antara car year dengan commission earned
print('Nilai korelasi antara tahun rilis mobil dengan tingkat komisi yang dibayarkan kepada penjual adalah:', data["Car_Year"].corr(data["Commission_Earned"]))
```
