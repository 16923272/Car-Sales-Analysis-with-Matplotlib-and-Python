```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel("car_sales_data.xlsx")
```

```python
# Bar Chart
data["Car_Model"].value_counts().plot(kind="bar",x="Car_Model",ylabel="Frekuensi",xlabel="Jenis Mobil",title="Grafik Batang Jumlah Pembeli Mobil sesuai dengan Jenisnya")
```

```python
# Line Chart
data.plot(kind="line", x="Date", y="Commission_Earned", title="Line Chart Commision Earned terhadap Date")
```

```python
# Pie Chart
data["Car_Make"].value_counts().plot(kind="pie",x="Merek Mobil",ylabel = "",title="Merek Mobil yang dijual pada tahun 2022-2023")
```

```python
# Scatter Plot
data.plot(kind="scatter", x="Sale_Price", y="Commission_Earned", xlabel="Harga Penjualan Mobil", ylabel="jumlah komisi yang diperoleh penjual", title="Diagram Tebar Harga Jual Mobil Dengan Jumlah Komisi Yang Didapatkan")
```
